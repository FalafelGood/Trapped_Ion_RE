## Ingredient 0: Imperfect Bell pairs

```python
# Make some imperfect Bell pairs:
s = """
H 0
CX 0 1
DEPOLARIZE1(0.1) 0
M 0 1
DETECTOR rec[-1] rec[-2]
"""
c = stim.Circuit(s)
c.diagram(type="timeline-svg")

def badBellPair(F=0.9):
    return f"""
    H 0
    CX 0 1
    DEPOLARIZE1({1-F}) 0
    """

c = stim.Circuit(badBellPair())
c.diagram(type="timeline-svg")
```

## Ingredient 1: CNOT gate teleportation
Source: "Deterministic teleportation of a quantum gate between 2 logical qubits"

```python
# Teleported CNOT circuit constructor
"""
s, t: source and target qubits
b0, b1: bell pair qubits
"""
def tCNOT(s, t, b0, b1):
    # Note rec[-1] is the most recent measurement
    # rec[-2] is second most recent, etc.
    s = f"""
    CNOT {s} {b0}
    CNOT {b1} {t}
    Z {b0} 
    Z {b1}
    M {b0}
    M {b1}
    CX rec[-2] {t}
    CZ rec[-1] {s}
    """
    return s

s = tCNOT(0, 3, 1, 2)
c = stim.Circuit(s)
c.diagram(type="timeline-svg")
```

## Ingredient 2: Parity check purification

Source: Various -- "Optimized entanglement purification"

```python
# Parity check purification circuit constructor:
"""
a0, a1: Bell pair qubits
b0, b1: ditto
"""
def pcheck(a0, a1, b0, b1):
    s = f"""
    CNOT {b0} {a0}
    CNOT {b1} {a1}
    M {a0}
    M {a1}
    """
    return s

s = pcheck(0, 3, 1, 2)
c = stim.Circuit(s)
c.diagram(type="timeline-svg")
```

Though Stim is generally quite fast for stabilizer simulation, Gidney [cautions](https://quantumcomputing.stackexchange.com/questions/22281/simulating-flag-qubits-and-conditional-branches-using-stim) that incorperating feedforward measurements is around 100x slower.

```python
"""
Parity check purification circuit construction with simulation. 

This function performs the following steps:

1. Given a number of comms ions and a probability of establishing end-to-end entanglement,
samples from the appropriate binomial distribution to determine how many Bell pairs we have
to purify

2. If we have more than 1 Bell pair, concatanate purification circuits for every 2 
Bell pairs

3. Construct a Stim simulator, and evaluate it against the circuit. This produces (among 
other things) a stabilizer Tableau for the circuit and a set of measurement outcomes

4. Note which purifications were successful and return the relevant data.

Inputs:
num_comms: Number of communication ions
p_e: Probability of establishing end-to-end entanglement
num_stored: Stockpiled pairs to include
F: Fidelity of the input Bell pairs

Returns:
circuit: The quantum circuit (for visualization purposes)
simulator: A Stim datatype containing the simulated circuit instance (including measurements)
purified_pairs: A list of qubits corresponding to the locations of successful purified bell pairs (ordered in pairs)
extra_pairs: 0 or 1
"""
def purification(num_comms=100, p_e=2.2e-4, num_stored=0, F=0.9):
    num_pairs = np.random.binomial(num_comms, p_e) + num_stored
    if num_pairs == 0 or num_pairs == 1:
        return stim.Circuit(), stim.TableauSimulator(stim.Circuit()), [], num_pairs
    
    extra_pairs = num_pairs % 2
    num_purifiers = int((num_pairs - extra_pairs)/2)
    circuit = stim.Circuit()
    qubits = range(4*num_purifiers)
    pair_tops = qubits[::2]
    purif_tops = qubits[::4] # top qubit of each purification circuit
    
    # Bell pair constructors (Unfortunately, these seperate loops are necessary to parallelize circuit ops)
    for q in purif_tops:
        circuit.append_operation("H", [q, q+1])
    for q in purif_tops:
        circuit.append_operation("CNOT", [q, q+3])
    for q in purif_tops:
        circuit.append_operation("CNOT", [q+1, q+2])
    for q in purif_tops:
        circuit.append_operation("DEPOLARIZE1", [q, q+1], [1-F])
        
    # Purification circuits
    for q in purif_tops:
        circuit.append_operation("CNOT", [q+1, q, q+2, q+3])
    for q in purif_tops:
        circuit.append_operation("M", [q, q+3])
        
    # Measure:
    simulator = stim.TableauSimulator()
    
    # Run first part of circuit
    simulator.do(circuit)
    results = simulator.current_measurement_record()
    # TODO: Check statistics. Does the probability change with Fidelity? Bennett seems to indicate yes, but need confirmation
    
    # Track succesful pairs
        # In principle, we only need to know the number of successful purifications.
        # In practice, there may be circuit noise. To be safe, I'll track the successful purifications and build the circuit around this.
    purified_pairs = []
    m_locs = range(2*num_purifiers) # measurement locations
    for m in m_locs[::2]:
        if results[m] == results[m+1]:
            purified_pairs.extend([2*m+1, 2*m+2]) # extend is a cleaner alternative to append
    
    return circuit, simulator, purified_pairs, extra_pairs
    
c, simulator, purified_pairs, extra_pairs = purification(num_comms=20, p_e=0.4, num_stored=0, F=0.9)
print(purified_pairs)
c.diagram(type="timeline-svg")
```

### Discussion: Is conditional Teleportation necessary?


In practice, we know which purifications are successful by measuring. With all-to-all connectivity we technically don't need conditional teleportation.

### Open question(s)

Q: We're given that the success probability of e2ee is low and measurement times are high. Is the success probability so low that it would be worth performing purification with measurement?

A: There's no advantage here. We'd still have to purify the last pairs to arrive and so we're rate limited in the same way that we'd be if we decided to push purifcation measurements to the end. The only hope of actually improving the rate is if we're able to perform the measurements for purification and conditional teleportation at the same time. Is this possible? We'd need to be able to retcon a Z measurement into an X or vice versa.

Actually, this is possible to some extent! You can condition teleportation after the measurement provided you have pure Bell pairs.

```python
"""
Part 2 of the simulation:

Here we perform conditional teleportation onto a number of ancilla registers.

How useful is this in practice?

returns:
simulator:
targets_met: A list of Bell pairs we were able to satisfy.
"""
def conditional_teleportation(pur_circuit, simulator, purified_pairs, num_targets):
    if purified_pairs == []:
        return pur_circuit, simulator, 0
    
    nq = pur_circuit.num_qubits
    qubits = range(nq)
    num_pairs = len(purified_pairs)/2
    tele_circuit = stim.Circuit()
    
    # Why is conditional teleportation necessary? Presumably because measurements are deferred?
    # In which case I might have done a lot of work for no reason...
    
    tele_circuit.append_operation("H", num_qubits)
    tele_circuit.append_operation("CNOT", [num_qubits, num_qubits+1])
    
    
    # update the simulation with conditional teleportation
    simulator.do(tele_circuit)
    
    # update the `master' circuit to include conditional teleportation
    pur_circuit.__iadd__(tele_circuit)
    
    return

c, simulator, purified_pairs, extra_pairs = purification(num_comms=20, p_e=0.4, num_stored=0, F=0.9)
conditional_teleportation(c, simulator, purified_pairs, 1)
c.diagram(type="timeline-svg")
```

```python
# Conditional Bell pair teleportation

# How useful is this actually?
"""
a0 a1: Bell pair qubits
b0 b1: ditto
c0 c1: Target register
exist_a: Does the bell pair at (a0, a1) exist?
exist_b: ditto
"""
def cond_bell_tele(a0, a1, b0, b1, c0, c1, exist_a: bool, exist_b: bool):
    if (exist_a and not exist_b) or (exist_a and exist_b):
        return f"""
        CNOT {a0} {c0}
        CNOT {a1} {c1}
        CNOT {b0} {c0}
        CNOT {b1} {c1}
        
        H {a0}
        H {a1}
        
        M {a0}
        M {a1}
        M {b0}
        M {b1}
        
        CX rec[-4] {c0}
        CX rec[-3] {c1}
        CZ rec[-2] {c0}
        CZ rec[-1] {c1}
        """
    elif exist_b and not exist_a:
        return f"""
        CNOT {a0} {c0}
        CNOT {a1} {c1}
        CNOT {b0} {c0}
        CNOT {b1} {c1}
        
        H {b0}
        H {b1}
        
        M {a0}
        M {a1}
        M {b0}
        M {b1}
        
        CX rec[-4] {c0}
        CX rec[-3] {c1}
        CZ rec[-2] {c0}
        CZ rec[-1] {c1}
        """
    else:
        return ""

s = cond_bell_tele(0, 1, 2, 3, 4, 5, False, True)
c = stim.Circuit(s)
c.diagram(type="timeline-svg")
```
