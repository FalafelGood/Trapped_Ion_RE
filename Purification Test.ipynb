```python
import stim
import numpy as np
import itertools
from copy import copy
from copy import deepcopy
import matplotlib.pyplot as plt
```

```python
# Construct simple purification circuit for different noise types:
# Options: X_ERROR, Y_ERROR, Z_ERROR, DEPOLARIZE1, ... 
# See https://github.com/quantumlib/Stim/blob/main/doc/gates.md for full list
def simple_purification_circuit(p = 0.0, noise_type="DEPOLARIZE1"):
    circuit = stim.Circuit()
    # Bell state preperation
    circuit.append("H", [0,1])
    circuit.append("CNOT", [0,3,1,2])
    if noise_type == "DEPOLARIZE2":
        circuit.append("DEPOLARIZE2", [0,3,1,2], p)
    else:
        circuit.append(noise_type, [0,1], p)
    circuit.append("CNOT", [1,0,2,3])
    # Error prone cloning and teleportation
    circuit.append("MZ", [0,3])
    circuit.append("R", [0,3])
    return circuit

circuit = simple_purification_circuit(0.0, "X_ERROR")
circuit.diagram(type="timeline-svg")
```

```python
circuit = simple_purification_circuit(0.1, "X_ERROR")
simulator = stim.TableauSimulator()
simulator.do(circuit)
print(simulator.current_measurement_record())
simulator.state_vector()
```

```python
ideal_sv = np.zeros(2 ** 4, complex)
ideal_sv[0] = 1/np.sqrt(2)
ideal_sv[6] = 1/np.sqrt(2)
print(ideal_sv)
```

```python
def ave_fidelity(num_trials, f_init, noise_type):
    ave_F = 0.0
    missed_counts = 0
    circuit = simple_purification_circuit(1-f_init, noise_type)
    for N in range(num_trials):
        simulator = stim.TableauSimulator() # important step, am I resetting where I need to?
        simulator.do(circuit)
        measurements = simulator.current_measurement_record()
        if measurements[0] == measurements[1]:
            sv = simulator.state_vector()
            ave_F += np.inner(sv, ideal_sv)
        else:
            missed_counts += 1
#     print(f"Missed counts = {missed_counts}")
    return ave_F / (num_trials - missed_counts)

ave_fidelity(100, 0.1, "Z_ERROR")
```

```python
def plot_ave_fidelity(num_trials, noise_type, show_plot=True):
    f_init = np.arange(0., 1., 0.02)
    Fs = []
    for f in f_init:
        Fs.append(ave_fidelity(num_trials, f, noise_type))
    
    if show_plot == True:
        # f, ax = plt.subplots(figsize=(5,5))
        plt.figure(figsize=(5,5))
        plt.plot(f_init, Fs)
        plt.xlabel("F init")
        plt.ylabel("F out")
        plt.show()
    
    return Fs
    
test = plot_ave_fidelity(1000, "Y_ERROR")
```

```python
# Plot fidelities for all error types of interest
def plot_all(num_trials):
    f_init = np.arange(0., 1., 0.02)
    error_types = ["X_ERROR", "Y_ERROR", "Z_ERROR", "DEPOLARIZE1", "DEPOLARIZE2"]
    all_fidelities = []
    for e in error_types:
        print(f"Collecting data for {e}")
        all_fidelities.append(plot_ave_fidelity(num_trials, e, show_plot=False))
    plt.figure(figsize=(5,5))
    for idx, f_out in enumerate(all_fidelities):
        plt.plot(f_init, f_out, label=error_types[idx])
    
    # Get data for analytic curves
    def Xdephasing(probXerror):
        Xfidelities = []
        for p in probXerror:
            Xfidelities.append(p**2/(p**2 + (1-p)**2))
        return Xfidelities
    
    def Depolarizing(probDepol):
        depolFidelities = []
        for p in probDepol:
            depolFidelities.append((p**2 + 1/9 * (1-p)**2) / (p**2 + 2/3 * p*(1-p) + 5/9 * (1-p)**2))
        return depolFidelities
    
    Xfidelities = Xdephasing(f_init)
    depolFidelities = Depolarizing(f_init)
    plt.plot(f_init, Xfidelities, label="Analytic X dephasing", linestyle="dotted", linewidth=3, color="r")
    plt.plot(f_init, depolFidelities, label="Analytic depolarizing", linestyle="dotted", linewidth=3, color="b")
    
    plt.xlabel("Probability of error")
    plt.ylabel("Output Fidelity")
    plt.legend()
    plt.show()

plot_all(10000)
```
